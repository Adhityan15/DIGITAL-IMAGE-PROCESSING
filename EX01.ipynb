```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('input4.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Split RGB channels
R, G, B = cv2.split(img_rgb)

# Convert to grayscale
gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

# Negative of grayscale image
negative_gray = 255 - gray

# Convert grayscale to binary
_, binary = cv2.threshold(gray, 127, 1, cv2.THRESH_BINARY)

# Negative of binary image
negative_binary = 1 - binary

# Display images
plt.figure(figsize=(12, 10))

plt.subplot(3, 3, 1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(3, 3, 2)
plt.title("Red Component")
plt.imshow(R)
plt.axis('off')

plt.subplot(3, 3, 3)
plt.title("Green Component")
plt.imshow(G)
plt.axis('off')

plt.subplot(3, 3, 4)
plt.title("Blue Component")
plt.imshow(B)
plt.axis('off')

plt.subplot(3, 3, 5)
plt.title("Grayscale Image")
plt.imshow(gray)
plt.axis('off')

plt.subplot(3, 3, 6)
plt.title("Negative Grayscale Image")
plt.imshow(negative_gray)
plt.axis('off')

plt.subplot(3, 3, 7)
plt.title("Binary Image")
plt.imshow(binary)
plt.axis('off')

plt.subplot(3, 3, 8)
plt.title("Negative Binary Image")
plt.imshow(negative_binary)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# =========================
# 1 Read Image
# =========================

img = cv2.imread('input4.png')

if img is None:
    print("Error: Image not found. Check file name.")
    exit()

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# =========================
# 2 Split RGB Channels
# =========================

R, G, B = cv2.split(img_rgb)

red_img = np.zeros_like(img_rgb)
red_img[:, :, 0] = R

green_img = np.zeros_like(img_rgb)
green_img[:, :, 1] = G

blue_img = np.zeros_like(img_rgb)
blue_img[:, :, 2] = B

# =========================
# 3 Grayscale
# =========================

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# =========================
# 4 Binary Image
# =========================

_, binary_img = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)

# =========================
# 5 Negative Images
# =========================

negative_color = 255 - img_rgb
negative_gray = 255 - gray_img
negative_binary = 255 - binary_img

# =========================
# 6 Display All Images
# =========================

plt.figure(figsize=(14, 10))

plt.subplot(3, 4, 1)
plt.title("Original Color")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(3, 4, 2)
plt.title("Red Channel")
plt.imshow(red_img)
plt.axis('off')

plt.subplot(3, 4, 3)
plt.title("Green Channel")
plt.imshow(green_img)
plt.axis('off')

plt.subplot(3, 4, 4)
plt.title("Blue Channel")
plt.imshow(blue_img)
plt.axis('off')

plt.subplot(3, 4, 5)
plt.title("Grayscale")
plt.imshow(gray_img, cmap='gray')
plt.axis('off')

plt.subplot(3, 4, 6)
plt.title("Binary")
plt.imshow(binary_img, cmap='gray')
plt.axis('off')

plt.subplot(3, 4, 7)
plt.title("Negative Color")
plt.imshow(negative_color)
plt.axis('off')

plt.subplot(3, 4, 8)
plt.title("Negative Gray")
plt.imshow(negative_gray, cmap='gray')
plt.axis('off')

plt.subplot(3, 4, 9)
plt.title("Negative Binary")
plt.imshow(negative_binary, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
