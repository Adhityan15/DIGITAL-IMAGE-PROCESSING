```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('input.png',0)   # grayscale


h,w = img.shape
cropped = img[h//4:h//2 , w//4:w//2]

# -------------------------------
# 3. Compute Mean
# -------------------------------

mean_val = np.mean(img)

# -------------------------------
# 4. Compute Standard Deviation
# -------------------------------

std_val = np.std(img)

# -------------------------------
# 5. Create Image1 (half black + checkerboard)
# -------------------------------

checker = np.zeros((200,200))

for i in range(200):
    for j in range(200):
        if (i//20 + j//20) % 2 == 0:
            checker[i,j] = 255

image1 = np.zeros((200,200))
image1[:,0:100] = checker[:,0:100]
image1[:,100:] = 0

# -------------------------------
# 6. Create Image2 (checkerboard)
# -------------------------------

image2 = checker

# -------------------------------
# 7. Correlation coefficient
# -------------------------------

X = image1.flatten()
Y = image2.flatten()

corr = np.corrcoef(X,Y)[0,1]

# -------------------------------
# 8. Print values
# -------------------------------

print("Mean =",mean_val)
print("Standard Deviation =",std_val)
print("Correlation Coefficient =",corr)

# -------------------------------
# 9. Display images
# -------------------------------

plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.title("Original Image")
plt.imshow(img,cmap='gray')

plt.subplot(2,2,2)
plt.title("Cropped Image")
plt.imshow(cropped,cmap='gray')

plt.subplot(2,2,3)
plt.title("Image1")
plt.imshow(image1,cmap='gray')

plt.subplot(2,2,4)
plt.title("Image2")
plt.imshow(image2,cmap='gray')

plt.show()


```
