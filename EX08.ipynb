```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 1 Read Image
# -----------------------------
img = cv2.imread('test.png')

# -----------------------------
# 2 Convert to Grayscale
# -----------------------------
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# -----------------------------
# 3 Add Noise (Salt & Pepper)
# -----------------------------
noise = gray.copy()

prob = 0.02
for i in range(noise.shape[0]):
    for j in range(noise.shape[1]):
        r = np.random.random()
        if r < prob:
            noise[i][j] = 0
        elif r > 1 - prob:
            noise[i][j] = 255

# -----------------------------
# 4 Mean Filtering
# -----------------------------
mean3 = cv2.blur(noise,(3,3))
mean10 = cv2.blur(noise,(10,10))

# -----------------------------
# 5 Median Filtering
# -----------------------------
median = cv2.medianBlur(noise,5)

# -----------------------------
# 6 Display Images
# -----------------------------
plt.figure(figsize=(10,8))

plt.subplot(2,3,1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

plt.subplot(2,3,2)
plt.title("Gray Image")
plt.imshow(gray,cmap='gray')

plt.subplot(2,3,3)
plt.title("Noise Added Image")
plt.imshow(noise,cmap='gray')

plt.subplot(2,3,4)
plt.title("3x3 Image")
plt.imshow(mean3,cmap='gray')

plt.subplot(2,3,5)
plt.title("10x10 Image")
plt.imshow(mean10,cmap='gray')

plt.subplot(2,3,6)
plt.title("Median Filter")
plt.imshow(median,cmap='gray')

plt.show()

# -----------------------------
# Second Example (like record)
# -----------------------------
img2 = cv2.imread('image2.jpg',0)

filtered1 = cv2.blur(img2,(5,5))
filtered2 = cv2.medianBlur(img2,5)

plt.figure(figsize=(8,4))

plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow(img2,cmap='gray')

plt.subplot(1,3,2)
plt.title("Filtered Image")
plt.imshow(filtered1,cmap='gray')

plt.subplot(1,3,3)
plt.title("Filtered Image 2")
plt.imshow(filtered2,cmap='gray')

plt.show()
```
