```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Step 1: Read Image (Grayscale)
# -----------------------------
img = cv2.imread("input9.jpg", cv2.IMREAD_GRAYSCALE)

if img is None:
    print("Error: Image not found!")
    exit()

plt.figure(figsize=(5,5))
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()


# ------------------------------------
# Step 2: 1-D FFT of a Selected Row
# ------------------------------------
row = img[img.shape[0] // 2, :]   # Middle row

fft_1d = np.fft.fft(row)
fft_1d_shifted = np.fft.fftshift(fft_1d)

magnitude_1d = np.abs(fft_1d_shifted)

plt.figure(figsize=(8,4))
plt.plot(magnitude_1d)
plt.title("1-D FFT Magnitude Spectrum (Middle Row)")
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.show()


# ------------------------------------
# Step 3: 2-D FFT of Image
# ------------------------------------
fft_2d = np.fft.fft2(img)
fft_2d_shifted = np.fft.fftshift(fft_2d)

# ------------------------------------
# Step 4: Magnitude Spectrum
# ------------------------------------
magnitude_2d = np.abs(fft_2d_shifted)
log_magnitude = np.log(1 + magnitude_2d)

plt.figure(figsize=(6,6))
plt.imshow(log_magnitude, cmap='gray')
plt.title("2-D FFT Magnitude Spectrum")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# --------------------------------
# Step 1: Read Grayscale Image
# --------------------------------
img = cv2.imread("OIP.jpg", cv2.IMREAD_GRAYSCALE)

if img is None:
    print("Error: Image not found!")
    exit()

# --------------------------------
# Step 2: Compute 2-D FFT
# --------------------------------
F = np.fft.fft2(img)

# --------------------------------
# Step 3: Center the Spectrum
# --------------------------------
F_shifted = np.fft.fftshift(F)

# --------------------------------
# Step 4: Compute Magnitude Spectrum
# --------------------------------
magnitude_spectrum = np.abs(F_shifted)

# Log Scaling for better visualization
log_spectrum = np.log(1 + magnitude_spectrum)

# --------------------------------
# Step 5: Display Results
# --------------------------------
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(np.abs(F), cmap='gray')
plt.title("2-D FFT Image")
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title("Centered Frequency Spectrum")
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(log_spectrum, cmap='gray')
plt.title("Log Magnitude Spectrum (log(1+|F|))")
plt.axis('off')

plt.tight_layout()
plt.show()
```
