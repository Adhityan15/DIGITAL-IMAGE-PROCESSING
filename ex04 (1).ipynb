```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read Color Image
img_color = cv2.imread('input6.jpg')

if img_color is None:
    print("ERROR: Image not found")
    exit()

# Convert BGR to RGB for matplotlib
img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

# Step 2: Convert to Grayscale
gray = cv2.cvtColor(img_color, cv2.COLOR_RGB2GRAY)

# Step 3: Contrast Stretching (Grayscale)
r_min = np.min(gray)
r_max = np.max(gray)

enhanced1 = ((gray - r_min) / (r_max - r_min)) * 255
enhanced1 = enhanced1.astype(np.uint8)

# Step 4: Histogram Equalization (Grayscale)
equalized_gray = cv2.equalizeHist(gray)

# Step 5: Histogram Equalization for Color Image
# Convert to YCrCb color space
img_ycrcb = cv2.cvtColor(img_color, cv2.COLOR_RGB2YCrCb)

# Equalize only Y (luminance) channel
img_ycrcb[:, :, 0] = cv2.equalizeHist(img_ycrcb[:, :, 0])

# Convert back to RGB
equalized_color = cv2.cvtColor(img_ycrcb, cv2.COLOR_YCrCb2RGB)

# Display Images
plt.figure(figsize=(15, 10))

plt.subplot(2,3,1)
plt.title("Original Color Image")
plt.imshow(img_color)
plt.axis('off')

plt.subplot(2,3,2)
plt.title("Grayscale Image")
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.subplot(2,3,3)
plt.title("Contrast Stretched (Gray)")
plt.imshow(enhanced1, cmap='gray')
plt.axis('off')

plt.subplot(2,3,4)
plt.title("Equalized (Gray)")
plt.imshow(equalized_gray, cmap='gray')
plt.axis('off')

plt.subplot(2,3,5)
plt.title("Equalized Color Image")
plt.imshow(equalized_color)
plt.axis('off')

plt.tight_layout()
plt.show()

# Display Histograms
plt.figure(figsize=(15,8))

plt.subplot(2,2,1)
plt.title("Original Histogram (Gray)")
plt.hist(gray.ravel(), bins=256, range=[0,256])

plt.subplot(2,2,2)
plt.title("Contrast Stretched Histogram")
plt.hist(enhanced1.ravel(), bins=256, range=[0,256])

plt.subplot(2,2,3)
plt.title("Equalized Histogram (Gray)")
plt.hist(equalized_gray.ravel(), bins=256, range=[0,256])


plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('input4.jpg', 0)  # 0 = grayscale
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Original Histogram")
plt.hist(img.ravel(), bins=256, range=[0,256])
plt.show()
```

```python
r_min = np.min(img)
r_max = np.max(img)

stretched = ((img - r_min) / (r_max - r_min)) * 255
stretched = stretched.astype(np.uint8)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Contrast Stretched Image")
plt.imshow(stretched, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Stretched Histogram")
plt.hist(stretched.ravel(), bins=256, range=[0,256])
plt.show()
```

```python
equalized = cv2.equalizeHist(img)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Equalized Image")
plt.imshow(equalized, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Equalized Histogram")
plt.hist(equalized.ravel(), bins=256, range=[0,256])
plt.show()
```
